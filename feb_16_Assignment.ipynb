```python
# Q1. What is a database? Differentiate between SQL and NoSQL databases.
# A database is a collection of data that is organized in a systematic manner to facilitate easy retrieval, storage,
# and management of information. It is an essential component of most modern applications and is used to store and manage data in a structured format.

# SQL and NoSQL are two different types of databases, each with its own set of advantages and disadvantages.

# SQL, or Structured Query Language, is a relational database management system that is used to store and manage data in a
# tabular format, with data organized into rows and columns. SQL databases are widely used in applications that require 
# structured data, such as financial systems, e-commerce platforms, and content management systems. SQL databases are characterized 
# by their support for transactions, strong consistency, and the ability to handle complex queries. Some popular SQL databases include
# MySQL, PostgreSQL, and Oracle.

# NoSQL, or "not only SQL", is a non-relational database management system that is used to store and manage unstructured data.
# NoSQL databases are designed to handle large volumes of data with high availability and scalability. They are ideal for applications
# that require fast access to large volumes of data, such as social networks, online gaming platforms, and big data applications.
# NoSQL databases are characterized by their support for eventual consistency, high availability, and horizontal scalability. 
# Some popular NoSQL databases include MongoDB, Cassandra, and Redis.
```

```python
# Q5. Explain Primary Key and Foreign Key.
# In a relational database, a primary key and a foreign key are two important concepts that help establish relationships between tables.

# A primary key is a column or a set of columns in a table that uniquely identifies each row in that table.
# The primary key is used to enforce data integrity, ensure uniqueness, and provide a way to reference the data in the table from other tables.
# Each table can have only one primary key.

# For example, consider a table called "employees" that has the following columns: "employee_id", "name", "email", and "department_id".
# The "employee_id" column can be used as the primary key for this table, as each employee will have a unique identifier.

# A foreign key is a column or a set of columns in one table that refers to the primary key in another table.
# The foreign key establishes a link between the two tables, allowing data to be shared between them. 
# A foreign key constraint ensures that data in the referencing table (also known as the child table)
# must match data in the referenced table (also known as the parent table).

# For example, consider a table called "departments" that has the following columns: "department_id" and "name". 
# If we add a column "department_id" to the "employees" table, we can use it as a foreign key to link the "employees" table
# to the "departments" table. This allows us to create a one-to-many relationship between the two tables, 
# where each employee belongs to one department, and each department can have many employees.

# The foreign key constraint ensures that the values in the "department_id" column of the "employees" table must
# match the values in the "department_id" column of the "departments" table. This ensures that data in the two tables 
# is consistent and prevents data from being added to the "employees" table that does not correspond to a valid department 
# in the "departments" table.

# In summary, a primary key is a unique identifier for a row in a table, while a foreign key establishes a link
# between two tables by referring to the primary key in another table. Together, these two concepts help establish
# relationships between tables in a relational database.
```
