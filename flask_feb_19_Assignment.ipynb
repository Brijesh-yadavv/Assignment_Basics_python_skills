```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?
# =>
# Flask is a lightweight, flexible, and popular Python web framework used to build web applications.
# It is known for its simplicity, ease of use, and its ability to provide a solid foundation for building scalable and secure web applications.
# Flask is open-source and is developed and maintained by a large community of developers.

# Advantages of Flask Framework:

# Lightweight and flexible: Flask is a lightweight framework that does not require any specific tools or libraries.
# It is flexible and can be easily customized as per the specific needs of the application.

# Simple and easy to learn: Flask has a simple and intuitive interface which makes it easy for developers to get started with it.
# Its minimalist design makes it ideal for building smaller projects or prototypes quickly.

# Scalable: Flask is designed to be modular and scalable, making it suitable for building complex web applications. 
# It can be easily extended using third-party libraries and tools.

# Built-in development server: Flask comes with a built-in development server, which makes it easy for developers to test their code and debug it.

# Jinja2 templating: Flask uses Jinja2 as its templating engine, which provides powerful features such as template inheritance, filters, and macros.

# Large and active community: Flask has a large and active community of developers who regularly contribute to the development and 
# maintenance of the framework. This community provides support, tutorials, and documentation for Flask.

# Support for extensions: Flask has a wide range of extensions available for various functionalities such as database connectivity, 
# authentication, and caching. These extensions can be easily integrated into Flask applications to add new features and functionalities.

# Overall, Flask is an excellent choice for building web applications due to its simplicity, flexibility, and scalability.
# Its active community and extensive documentation make it easy for developers to get started and build high-quality web applications quickly.
```
