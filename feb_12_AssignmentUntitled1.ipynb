```python
# what is an Exception in python and what is difference between Exception and Syntax Error?
# =>
# An exception in Python is an error that occurs during the execution of a Python program.
# When an exception occurs, the program stops executing and raises an error message, indicating the type of exception 
# and the line of code where the exception occurred.

# The main difference between exceptions and syntax errors is that syntax errors occur during the parsing of the program,
# before any code is executed, while exceptions occur during the execution of the program.

# Here are some key differences between exceptions and syntax errors:

# Syntax errors occur due to mistakes in the program's syntax, such as missing parentheses or quotes, while exceptions occur
# due to errors in the program's logic or runtime conditions.

# Syntax errors prevent the program from running at all, while exceptions can occur during the execution of the program and may or
# may not stop the program depending on how they are handled.

# Syntax errors are easy to detect as they are raised immediately when the code is compiled, while exceptions may require more effort to detect 
# and handle appropriately.

# In summary, exceptions are errors that occur during the execution of a Python program, while syntax errors occur during the compilation of the 
# program due to mistakes in the program's syntax.


```
